```python
import datetime

start = datetime.datetime.now()
print('Script started:')
print(start)

import pandas as pd
import folium
from folium.plugins import MarkerCluster, MeasureControl, HeatMap, HeatMapWithTime




#Create Map
m = folium.Map(location=[57, -4], zoom_start=7, tiles='cartodbpositron')
folium.TileLayer('stamenterrain').add_to(m)
```

```python
scotch = pd.read_csv("whisky.csv")
```

```python
scotch.Region.unique()
```

```python
Locations = scotch[['Lat', 'Lon']]
Location_List = Locations.values.tolist()
```

```python
dist = folium.FeatureGroup(name='Distilleries')
for point in range(0, len(Location_List)):
    folium.Marker(Location_List[point],  
                  popup=scotch['Distillery'][point],
                  icon=folium.Icon(color="blue", icon='industry', prefix='fa')
                  ).add_to(dist)
```

```python
highland = scotch[(scotch.Region == "Highland")]
speyside = scotch[(scotch.Region == "Speyside")]
islay_islands = scotch[(scotch.Region == "Islay & Islands")]
lowland = scotch[(scotch.Region == "Lowland")]
```

```python
Locations = highland[['Lat', 'Lon']]
Location_List = Locations.values.tolist()
```

```python
high = folium.FeatureGroup(name='Highland Distilleries')
for point in range(0, len(Location_List)):
    folium.Marker(Location_List[point],  
                  icon=folium.Icon(color="yellow", icon='industry', prefix='fa')
                  ).add_to(high)
```

```python
H_Locations = highland[['Lat', 'Lon']]
H_Location_List = H_Locations.values.tolist()
```

```python
dist.add_to(m)
high.add_to(m)

folium.LayerControl(collapsed=False).add_to(m)
m.add_child(MeasureControl())

m.save('scotch.html')

print('Map Created')

end = datetime.datetime.now()
print('Script completed:')
print(end)

print('Total time elapsed:')
print(end - start)
```
